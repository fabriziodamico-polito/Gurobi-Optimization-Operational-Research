## TV Series Scheduling

You have already finished studying and preparing yourself for tomorrow's exam. Having two hours and a half of spare time, you decide to watch some episodes of your favorite TV series, listed in Table 1. Since you get bored easily, you usually watch at most one episode of a given TV series, then you switch to another one

- Write the variables and parameters of the problem
- Write the mathematical model of the problem

| TV Series | Duration (min) |
|---|---|
| La casa de papel | 60 |
| Numb3rs | 40 |
| Stranger Things | 45 |
| The Defenders | 40 |


**INDEXES**:
- $i$ : series

**PARAMETERS**:
- $d_i$ : duration (min) of series i
- $T$ : max spare time

**VARIABLES**:
- $x_i$ : 1 if series i is watched, 0 otherwise

$$ \max \sum_i x_i $$
$$ s.t. \sum_i x_i d_i \le T $$

```python
import gurobipy as gp
from gurobipy import GRB

m = gp.Model("netflix")

series = ["La casa de papel", "Numb3rs", "Stranger Things", "The Defenders"]
durations = {"La casa de papel": 60, "Numb3rs": 40, "Stranger Things": 45, "The Defenders": 40}
T = 150

x = m.addVars(series, vtype=GRB.BINARY, name="x")

m.addConstr(gp.quicksum(x[i] * durations[i] for i in series) <= T, "TimeLimit")
m.setObjective(gp.quicksum(x[i] for i in series), GRB.MAXIMIZE)

m.optimize()

if m.status == GRB.OPTIMAL:
    print(f"Optimal number of episodes to watch: {int(m.objVal)}")
    print("Series to watch:")
    for i in series:
        if x[i].x > 0.5:
            print(f"- {i} ({durations[i]} min)")
```
