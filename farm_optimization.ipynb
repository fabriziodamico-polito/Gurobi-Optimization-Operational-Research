## Farm Production Planning

A farmer has 12 hectares of land available for growing lettuce or potatoes. 

The resources at his disposal, besides the land, are: 
- 70 kg of lettuce seeds
- 18 tons of tubers
- 160 cubic meters (m3) of fertilizer

Assuming that the market can absorb all the production and that prices are stable.
The estimated yield are:
- 3000 euros/hectare for lettuce cultivation
- 5000 euros/hectare for potatoes

The resource absorption for each type of cultivation is
-  7 kg of seeds and 10 m3 of fertilizer per hectare of lettuce
- 3 tons of tubers and 20 m3 of fertilizer per hectare of potatoes.

Determine how much land to allocate to lettuce and how much to potatoes in order to maximize the economic yield while making the best use of the available resources.

**INDEXES**:
- $L$ : lettuce
- $p$ : potatoes
- $f$ : fertilizer
- $t$ : tons

**PARAMETERS**:
- $y_L$ : yield for hectar of  lettuce 
- $y_p$ : yield for hectar of potatoes
- $M_L$ : maximum kg of lectuce seed
- $M_p$ : maximum tons of potatoes
- $M_f$ : maximum m3 of fertilizer
- $H$ : hectars available
- $a_{s,L}$ : seed absorption for hectar of lettuce
- $a_{f,L}$ : fertilizier absorption for hectar of lettuce
- $a_{t,p}$ : absorption of tons of potatoes per hectar
- $a_{f,p}$ : absorption of fertilizer for hectar of potatoes

**VARIABLES**:
- $x_L$ : hectars of lettuce
- $x_p$ : hectars of potatos

$$max \ \ \ x_L y_L + x_p y_p$$

$$ s.t \ \ x_L + y_L \le H$$
$$ x_L a_{s,L} \le M_L$$
$$ x_p a_{t,p} \le M_p$$
$$ x_L a_{f,L} + x_p a_{f,p} \le M_f$$
$$ x_L , x_p \ge 0$$

```python
import gurobipy as gp
from gurobipy import GRB

m = gp.Model("farm_optimization")

# Variables 
x_L = m.addVar(name = "x_L") # by default the variable is: greater than o (lb = 0), continuous (vtype = GRB.CONTINUOUS)
x_p = m.addVar(name = "x_p")

# Parameters
yields = {"L":3000, "p": 5000}
upper_bound = {"M_L":70, "M_p":18, "M_f":160, "H":12}
absorption = {"a_s_L":7, "a_f_L":10, "a_t_p":3, "a_f_p":20}

# Constraints
m.addConstr(x_L + x_p <= upper_bound["H"])
m.addConstr(x_L*absorption["a_s_L"]<= upper_bound["M_L"])
m.addConstr(x_p*absorption["a_t_p"]<=upper_bound["M_p"])
m.addConstr(x_L*absorption["a_f_L"]+x_p*absorption["a_f_p"]<=upper_bound["M_f"])

# Objective Funtion
m.setObjective(x_L*yields["L"] + x_p*yields["p"], GRB.MAXIMIZE)

m.optimize()

#output the results
print("-" * 30)
if m.status == GRB.OPTIMAL:
    print(f"Optimal Solution found")
    print(f"Maximum total profit: {m.objVal} €")
    print(f"- Hectars to Letuce (x_L): {x_L.x} ha")
    print(f"- Hectars to potatoes (x_p): {x_p.x} ha")
else:
    print("No optimal solution found")
```
