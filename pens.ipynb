## Pen Manufacturing Product Mix

A firm produces red and blue pens, whose unit production cost is 15 cents (including labor and raw materials). The firm incurs a daily fixed cost, amounting to 1000 €, to run the plant, which can produce at most 8000 pens per day in total (i.e., including both types). The blue pens sell for 25 cents, whereas things are a tad more complicated for the red ones. On a daily basis, the first 5000 red pens can be sold for 30 cents each, but the additional ones can be sold for only 20 cents.

- Write the variables and parameters of the problem
- Write the mathematical model of the problem
- The production manager decides to produce 500 red and 3000 blue pens. What is the value of this solution?
- Make a picture of the feasible space on the continuous variables.


Indexes:

b = blue pen
r = red pen

Parameters:

production cost = 15 cent
daily fixed cost = 1000 euro
maximum total production per day = 8000

sell price blu pen = 25 cent
sell price red pen (first 5000) = 30 cent
sell price red pen (additional) = 20 cent

Variables:

xb = quantity of blue pen producted
xr,1 = quantity of red pen until 5000
xr,2 = quantity of red pen over 5000

$$ max \ \ 0,25 x_b + 0,30 x_{r,1} + 0,20 x{r,2} - 0,15 (x_b + x_{r,1} + x_{r,2}) - 1000  $$
$$ s.t \ \ x_b + x_{r,1} + x_{r,2} \leq 8000 \\ x_{r,1} \leq 5000 $$


```python
import gurobipy as gp
from gurobipy import GRB

m = gp.Model("pens")
m.setParam("OutputFlag", 0)

x_b = m.addVar(vtype=GRB.CONTINUOUS, name="x_b")
x_r1 = m.addVar(vtype=GRB.CONTINUOUS, name="x_r1")
x_r2 = m.addVar(vtype=GRB.CONTINUOUS, name="x_r2")

m.addConstr(x_b + x_r1 + x_r2 <= 8000, "MaxProduction")
m.addConstr(x_r1 <= 5000, "RedThreshold")

m.setObjective(0.25 * x_b + 0.30 * x_r1 + 0.20 * x_r2 - 0.15 * (x_b + x_r1 + x_r2) - 1000, GRB.MAXIMIZE)
m.optimize()

if m.status == GRB.OPTIMAL:
    print(f"Optimal daily profit: {m.objVal:.2f} €")
    print(f"Blue pens to produce: {x_b.x}")
    print(f"Red pens to produce: {x_r1.x + x_r2.x} (first 5000: {x_r1.x}, additional: {x_r2.x})")
```
