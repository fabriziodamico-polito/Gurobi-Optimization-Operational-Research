A manufacturing company produces two types of pens (blue and red) with different selling prices and a shared daily production limit. Red pens have a tiered pricing structure depending on the quantity sold. Given production costs and a daily fixed cost, write a MILP to determine the optimal production mix to maximize the total daily profit.



Indexes:

b = blue pen
r = red pen

Parameters:

production cost = 15 cent
daily fixed cost = 1000 euro
maximum total production per day = 8000

sell price blu pen = 25 cent
sell price red pen (first 5000) = 30 cent
sell price red pen (additional) = 20 cent

Variables:

xb = quantity of blue pen producted
xr,1 = quantity of red pen until 5000
xr,2 = quantity of red pen over 5000

$$ max \ \ 0,25 x_b + 0,30 x_{r,1} + 0,20 x{r,2} - 0,15 (x_b + x_{r,1} + x_{r,2}) - 1000  $$
$$ s.t \ \ x_b + x_{r,1} + x_{r,2} \leq 8000 \\ x_{r,1} \leq 5000 $$
